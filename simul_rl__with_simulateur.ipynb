```python
sumoBinary = r"C:/Program Files/rl_project/Eclipse/Sumo/bin/sumo-gui.exe" 
sumoConfig = r"C:/Program Files/rl_project/2025-02-12-16-24-37/osm.sumocfg"
sumoConfig2 = r"C:/Program Files/rl_project/2025-02-25-19-05-44/osm.sumocfg"
# sumoConfig3 = r"/home/psels/code/psels/RL_traffic/RL_traffic/Traditional_traffic/traditional_traffic.sumo.cfg"
sumoConfig3 = r"Traditional_traffic/traditional_traffic.sumo.cfg"


sumoCmd = [sumoBinary, "-c", sumoConfig3, "--start"]


traci.start(sumoCmd)

trafic_light_ids = traci.trafficlight.getIDList()
# print("Les feux:", trafic_light_ids)



if not trafic_light_ids:
    print("No traffic lights, issues")
    traci.close()
    sys.exit()


lane_ids = traci.lane.getIDList()
print("lane id", lane_ids, len(lane_ids))

# Ini
step = 0
lane_situation = queue(lane_ids)

try:
    while step < 10000:


        # hard coded policy


        #grouping of traffic lights
        phase_to_change_number = [0, 2, 4, 6]
        combo_0 =['end1_junction_1','end3_junction_1']   # on va sommer le nombre de voiture pour ces deux lanes, si ce total est le plus grand, on passe à phase 0
        combo_2 = ['end1_junction_2','end3_junction_2']  # on va sommer le nombre de voiture pour ces deux lanes, si ce total est le plus grand, on passe à phase 2 etc..
        combo_4 = ['end2_junction_1','end4_junction_1']
        combo_6 = ['end4_junction_2', 'end2_junction_2']

        # #compputing the number of waiting cars per grouping of traffic lights
        car_combo_0 = sum([traci.lane.getLastStepHaltingNumber(i)for i in combo_0])
        car_combo_2 = sum([traci.lane.getLastStepHaltingNumber(i)for i in combo_2])
        car_combo_4 = sum([traci.lane.getLastStepHaltingNumber(i)for i in combo_4])
        car_combo_6 = sum([traci.lane.getLastStepHaltingNumber(i)for i in combo_6])
        values = [car_combo_0, car_combo_2, car_combo_4, car_combo_6 ]


        #selecting the phase to select according to the number of waiting cars per grouping traffic lights
        print("values", values)
        max_values_position = values.index(max(values))
        phase_to_change = phase_to_change_number[max_values_position]
        print("phase to change", phase_to_change)


        #setting the phase
        if step%1000==0:
            traci.trafficlight.setPhase(trafic_light_ids[0],phase_to_change) 

        traci.simulationStep()

        print('phase', traci.trafficlight.getPhase(trafic_light_ids[0]))
    

       
        step += 1

        

except Exception as e:
    print(f"exception with: {e}")

finally:
    traci.close()
    print("end")
```
